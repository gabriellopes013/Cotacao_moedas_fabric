```python
#Imports
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark.sql.types import *
import requests
import json
```

```python
url = "https://olinda.bcb.gov.br/olinda/servico/PTAX/versao/v1/odata/Moedas?top=100&top=100&format=json&$select=simbolo,nomeFormatado"

response = requests.get(url).json()['value']

df = spark.createDataFrame(response)

# display(df)

df = df.selectExpr(
    "nomeFormatado AS MoedaNome",
    "simbolo AS Moeda"
)

df.write.mode('overwrite').saveAsTable("dim_moedas")
```
